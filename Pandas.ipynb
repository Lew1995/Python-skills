```python
import pandas as pd

dataframe = pd.read_excel("/content/drive/MyDrive/Python/student.xlsx")
dataframe.info()

#select a column
dataframe["name"]  # returns a Series = 1 column
print(type(dataframe["name"]))

# Select multiple columns
dataframe[["id", "name"]] # returns a DataFrame

# Select n first rows of a dataframe
dataframe.head(1)

# Select n last rows of a dataframe (n = 5 by default)
dataframe.tail()
```

```python
# Slicing
my_list = ["Micayla", "Mehul", "Jack", "Sara", "Tooba", "Kolyu", "Patrick"]
print(my_list[0:2]) # end index is exclusive
print(my_list[0:5:2]) # start:stop:steps
print(my_list[2:]) # start at index 2 until end of list
print(my_list[:3]) # until index 3 non included
```

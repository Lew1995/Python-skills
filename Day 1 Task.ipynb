```python
# IF STATEMENT
animals = ["dog", "bird", "cat", "dragon", "cat"]


for animal in animals:
  if animal == "cat":       #more or condition met
    print("This was a cat!")
  elif animal == "dragon":  #same
    print("Wroaurrrrr")
  else:                     #less or everything else
    print("This is not a cat.")
```

```python
# Match
condition = "snake"
match condition:
  case "dog":
    print("This is not a cat!")
  case "cat":
    print("This IS a cat")
  case _:
    print("This is default behaviour")
```

```python
condition = 10

while condition > 0:
  print("Condition has been met")
  condition = condition - 1
  print(f"Condition value is now {condition}.")
```

```python
Count = list(range(100))

for count in Count:
  if count % 3 == 0 and count % 5 == 0:
    print("fizzbuzz")
  elif count % 3 == 0:
    print("fizz")
  elif count % 5 == 0:
    print("buzz")
  else:
    print(count)
```

```python
# Dictonary example
bearded_dragon = {
    "species": "reptile",
    "legs": 4,
    "favorite_food": "crickets",
    "likes_heat": True
}

unsatified_cat = {
    "species": "mammal",
    "legs": 4,
    "favorite_food": "uncertain",
}

animals = [bearded_dragon, unsatified_cat]
for animal in animals:
  print(animal)
  for key, value in animal.items():
    print(f"{key.capitalize()}: {value}")
```

```python
import random

user_input = int(input("I have picked a number between 10 and 99, can you guess what it is? "))
number_to_guess = random.randint(10, 99)
while user_input != number_to_guess:
    if user_input < number_to_guess:
          user_input = int(input("Little higher"))
    else:
         user_input = int(input("Little lower"))


print("Correct! You guessed it!")
```
